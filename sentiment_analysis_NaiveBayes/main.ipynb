```python
import pandas as pd
import matplotlib.pyplot as plt
import Sastrawi
import numpy as np
from wordcloud import WordCloud
```

CRAWLING DATA

```python
df = pd.read_csv('data_real.csv', index_col=0)
df.head()
```

```python
df_copy = df.copy()
```

LABELLING

```python
label = []
for index, row in df_copy.iterrows():
    if row['Rating'] == 1 or row['Rating'] == 2:
        label.append(0)
    else:
        label.append(1)

df_copy["label"] = label
```

```python
df_copy.head()
```

PREPROCESSING TEXT

```python
def hurufKecil(Review):
    Review = Review.lower()
    return Review

df_copy['Ulasan'] = df_copy['Ulasan'].apply(hurufKecil)
df_copy.head()
```

```python
# NORMALISASI

norm = {
    " dgn ": " dengan ",
    " tpi ": " tapi ",
    " tmpat ": " tempat ",
    " gx ": " gak ",
    " tpi ": " tapi ",
    " brg ": " barang ",
    " bgmn ": " bagaimana ",
    " hrs ": " harus ",
    " gk ": " gak ",
    " tdk ": " tidak ",
    " blm ": " belum ",
    " bgs ": " bagus ",
    " bgus ": " bagus ",
    " gtu ": " gitu ",
    " sy ": " saya ",
    " d ": " di ",
    " diseskripsi ": " di deskripsi ",
    " yg ": " yang "
}

def normalisasi(str_text):
    for i in norm:
        str_text = str_text.replace(i, norm[i])
    return str_text

df_copy['Ulasan'] = df_copy['Ulasan'].apply(lambda x: normalisasi(x))
```

```python
df_copy.head()
```

```python
# STOPWORD

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary 
# opsional jika ingin menambah kata baru
moreStopWords = []

stop_words = StopWordRemoverFactory().get_stop_words()
new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

def stopwords (str_text):
    str_text = stop_words_remover_new.remove(str_text)
    return str_text

df_copy['Ulasan'] = df_copy['Ulasan'].apply(lambda x: stopwords(x))
df_copy.head()
```

```python
# TOKENIZE

tokenized = df_copy['Ulasan'].apply(lambda x: x.split())
tokenized
```
